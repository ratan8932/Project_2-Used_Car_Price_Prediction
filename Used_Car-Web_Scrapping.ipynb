```python
pages = np.arange(1,55,1)
carname = []
make = []
model = []
year = []
fuelType = []
kilometerDriven = []
listingPrice = []
originalPrice = []
carId = []
city = []
mainImage = []
variant=[]
cashDownPayment=[]
dealRating=[]
exShowroomPrice=[]
onRoadPrice=[]
calculatedScore=[]
tofDiscount=[]
ownerNumber=[]
detail=[]
vehicleType=[]
city_id=[2920,2130,6356,290,2,134,5,2378,2423,4709,3686,5732,1692,132]
pin_id=[450001,302001,682001,226001,110001,201301,122001,400001,411001,560001,500001,600001,380001,201001]
for j,k in zip(city_id,pin_id):
    for page in pages:
        page = requests.get('https://api-sell24.cars24.team/buy-used-car?sort=P&serveWarrantyCount=false&gaId=
                            1947393616.1668261831&page='+str(page)+'&storeCityId={}&pinId={}'.format(j,k))
        result = page.json()
        for res in result["data"]["content"]:
            keys = res.keys()
            carname.append(res["carName"])
            make.append(res["make"])
            model.append(res["model"])
            year.append(res["year"])
            fuelType.append(res["fuelType"])
            kilometerDriven.append(res["kilometerDriven"])
            listingPrice.append(res["listingPrice"])
            originalPrice.append(res["originalPrice"])
            carId.append(res["carId"])
            city.append(res["city"])
            variant.append(res["variant"])
            cashDownPayment.append((res["cashDownPayment"]))
            dealRating.append(res["dealRating"])
            exShowroomPrice.append(res["exShowroomPrice"])
            onRoadPrice.append(res["onRoadPrice"])
            tofDiscount.append(res["tofDiscount"])
            calculatedScore.append(res["calculatedScore"])
            ownerNumber.append(res["ownerNumber"])
source1 = pd.DataFrame({'City':city,'Car_Name': carname, 'Maker': make,'Variant':variant,'Year' : year,'Model' : model,
                        'Fuel_Type' : fuelType,'KM_Driven' : kilometerDriven,'ownerNumber':ownerNumber,'Listing_Price' :
                        listingPrice,'Discount':tofDiscount,'Original_Price' : originalPrice,'Car_Id' : carId,
                        'Cash_Down_Payment': cashDownPayment,'Ex_Showroom_Price':exShowroomPrice,
                        'On_Road_Price':onRoadPrice,'CalculatedScore':calculatedScore})
source1
```

```python
pages = np.arange(1,16,1)
carname = []
make = []
model = []
year = []
fuelType = []
kilometerDriven = []
listingPrice = []
originalPrice = []
carId = []
city = []
mainImage = []
variant=[]
cashDownPayment=[]
dealRating=[]
exShowroomPrice=[]
onRoadPrice=[]
calculatedScore=[]
tofDiscount=[]
ownerNumber=[]
detail=[]
vehicleType=[]
for page in pages:
    page = requests.get('https://api-sell24.cars24.team/buy-used-car?sort=P&serveWarrantyCount=false&gaId=1947393616.1668261831&page='+str(page)+'&storeCityId=777&pinId=700001')
    result = page.json()
    for res in result["data"]["content"]:
        keys = res.keys()
        carname.append(res["carName"])
        make.append(res["make"])
        model.append(res["model"])
        year.append(res["year"])
        fuelType.append(res["fuelType"])
        kilometerDriven.append(res["kilometerDriven"])
        listingPrice.append(res["listingPrice"])
        originalPrice.append(res["originalPrice"])
        carId.append(res["carId"])
        city.append(res["city"])
        variant.append(res["variant"])
        cashDownPayment.append((res["cashDownPayment"]))
        dealRating.append(res["dealRating"])
        exShowroomPrice.append(res["exShowroomPrice"])
        onRoadPrice.append(res["onRoadPrice"])
        tofDiscount.append(res["tofDiscount"])
        calculatedScore.append(res["calculatedScore"])
        ownerNumber.append(res["ownerNumber"])
source = pd.DataFrame({'City':city,'Car_Name': carname, 'Maker': make,'Variant':variant,'Year' : year,'Model' : model,'Fuel_Type' : fuelType,'KM_Driven' : kilometerDriven,'ownerNumber':ownerNumber,'Listing_Price' :listingPrice,'Discount':tofDiscount,'Original_Price' : originalPrice,'Car_Id' : carId,'Cash_Down_Payment': cashDownPayment,'Ex_Showroom_Price':exShowroomPrice,'On_Road_Price':onRoadPrice,'CalculatedScore':calculatedScore})
source
```
